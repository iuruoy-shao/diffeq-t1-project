```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
image = np.asarray(Image.open('test2.jpeg'), dtype=float)
```

```python
image = image[0,:,:].T
plt.plot(list(np.arange(0, 180)), image[0], 'r')
plt.plot(list(np.arange(0, 180)), image[1], 'g')
plt.plot(list(np.arange(0, 180)), image[2], 'b')
plt.ylabel("Color Intensity")
plt.xlabel("Pixel Index")
```
